```python
import u3
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

d = u3.U3()
def set_dac_voltage(channel, bits):
    if channel == 0:
        d.getFeedback(u3.DAC0_8(bits))
    elif channel == 1:
        d.getFeedback(u3.DAC1_8(bits))
    else:
        print('Invalid channel number. Choose either 0 or 1.')
        return
    voltage_volts = bits / 256 * 4.95
    
    print(f'Voltage on DAC channel {channel}: {voltage_volts}')

channel = 0
#bits = 100 #bits i should loop bc all its giving me is the voltage on the oscilliscope
list = np.arange(10000)

for i in list:
    fint = 127*np.sin((i/100)*2*np.pi)+128
    integer = int(np.floor(fint))
    set_dac_voltage(0, integer)

#voltage_volts = bits / 256 * 4.95

#frequency = .5 #.5
#duration = 10
#sampling_rate = 1000
# list = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)


fint = 128*np.sin((list/100)*2*np.pi)+128
integer = [int(np.floor(f)) for f in fint]

y = integer
print(np.max(y))

#d.getFeedback(u3.DAC0_16, bits.astype(np.uint16))

plt.plot(list, y)
plt.xlabel('Loop Iterator')
plt.ylabel('Integer')
plt.title('Analog Output')
#plt.savefig('sine_wave.png')
plt.show()

d.close()
```
